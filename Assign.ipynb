### Logistic Regression for Bhuavana Data - EDA

```python
df_train = pd.read_csv("BhuvTrain.csv", low_memory=False)
df_train.head()

# plt.plot(df_train['Timestamp'], df_train['EDA'])
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
## Write code here to train a model using logreg.fit() which takes in train set features (as array)
## and the corresponding labels.
#logreg_model = 
data = df_train[['EDA','Timestamp']]
logreg_model = logreg.fit(np.array(data, dtype=pd.Series),df_train.Drunk )
#type(df_train.Drunk)
```

```python
df_test = pd.read_csv("BhuvTestLabelled.csv", low_memory=False)
test_data = df_test[['EDA','Timestamp']]
logreg_predictions = logreg_model.predict(test_data)
```

```python
accuracy_score(df_test.Drunk, logreg_predictions)
```

```python
df_test = pd.read_csv("astika-still-drunk/EDA.csv", low_memory=False)
test_data = df_test[['EDA','Timestamp']]
logreg_predictions = logreg_model.predict(test_data)
```

```python
accuracy_score(df_test.Drunk, logreg_predictions)
```
